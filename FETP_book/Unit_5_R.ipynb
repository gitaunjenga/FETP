- DataFrames allow you to group data based on certain columns and perform operations or summaries on each group. This functionality helps in analyzing data and generating meaningful insights from grouped information.



- Suppose you want to group the dataset by 'Referred by'.

```r
# count records by 'Referred by'

df %>%
  group_by(`Referred by`) %>%
  summarise(Counts = n())
```

#### 4.2. Data Quality Checks

Data quality checks are essential to ensure that your dataset is accurate, complete, and consistent. Performing these checks helps in identifying and correcting issues such as missing values, duplicate records, and inconsistencies.



> Missing values

The isnull method combined with sum helps us identify columns with missing values, which need to be addressed before further analysis.

```r
sapply(df, function(x) sum(is.na(x)))
```

> Duplicates

Duplicate records can skew your analysis results. Use the following methods to find and handle duplicates:

```r
duplicated_rows <- sum(duplicated(df))
print(paste('Number of duplicate rows:', duplicated_rows))
```

> Inconsistency

- Identifying and addressing inconsistencies in your dataset is crucial for maintaining data quality and ensuring accurate analysis. Logical inconsistencies, such as negative values in columns where they are not expected, can indicate errors or data issues.



```r
# logical inconsistencies

negative_age_records <- df %>%
  filter(`Age (years)` < 0)

# Display the number of records with Age (years) < 0
num_negative_age_records <- nrow(negative_age_records)
print(paste("Number of records with Age (years) < 0:", num_negative_age_records))
```

```r
# Display unique values in the Sex column

unique_sex_values <- unique(df$Sex)
print(unique_sex_values)
```

```r
library(dplyr)

# Filter rows where the date is in the future
future_dates_df <- df %>%
  filter(`Date seen at Health Facility/Community` > Sys.Date())

# Print the number of rows with future dates
cat("Number of rows with future dates:", nrow(future_dates_df), "\n")

# Display the rows with future dates
print(future_dates_df)
```

#### 4.3. Handling Data Quality Issues

Before merging or analyzing datasets, it's essential to clean and prepare them. This step ensures data integrity and consistency.



We can handle missing values by either filling them with appropriate values or dropping the rows/columns with missing values.



> Fill missing values with a specific value

```r
df <- df %>%
  mutate(`Residential District` = ifelse(is.na(`Residential District`), 'Missing', `Residential District`))
head(df)
```

> Removing Duplicates

```r
df <- df %>%
  distinct()
head(df)
```

### 5. Data Manipulation

> #### 5.1. Creating new columns

- Creating flags or derived variables in a DataFrame involves generating new columns based on existing data to facilitate analysis and insights. This process often includes converting data types and extracting specific components to create meaningful indicators, periods etc.


```r
# first we convert the specific column to the date type
library(lubridate)

df$`Date seen at Health Facility/Community` <- as.Date(df$`Date seen at Health Facility/Community`, format="%Y-%m-%d")
df <- df %>%
  mutate(Year_seen_at_facility = year(`Date seen at Health Facility/Community`))
head(df)
```

> #### 5.2. Filtering

- Filtering data allows you to extract subsets of data that meet specific criteria. This process helps in focusing on relevant records and performing targeted operations.

- To filter rows where the 'Residential Village/Suburb/Town' column contains the substring 'Kenema City', use the str.contains() method. This method checks if a string contains a particular substring and can handle missing values.

```r
# Filter rows where 'Residential Village/Suburb/Town' contains 'Kenema City'

filtered_df <- df %>%
  filter(grepl('Kenema City', `Residential Village/Suburb/Town`))
head(filtered_df)
```

- The filter() function subsets the rows of df where the condition inside is TRUE. The condition here is grepl('Kenema City', Residential Village/Suburb/Town). This checks each entry in the column Residential Village/Suburb/Town to see if it contains the substring "Kenema City".

> #### **5.3. Renaming columns**

Renaming columns to meaningful names enhances readability and consistency.

```r
df <- df %>%
  rename(
    Disease = `Disease/Condition/Event Notified`,
    Age = `Age (years)`
  )
head(df)
```

### 6. Meging datasets

 **6.1. Why merge datasets?**

- Merging datasets combines two or more different datasets, creating a “wider” combined dataset.

- Example: clinical data and laboratory data about the same case

- Data used in outbreak investigation rarely comes from a single source.  

- Each data source may store its data in its own database. To conduct your analysis, data from these sources must be stored in a single database.

- Some data sources may routinely exchange data with other software applications within the health information system. Their aggregate data may be stored in a single database or data warehouse.

This means that both datasets must:

- Have clean data
- Share the same structure, format, and organization
- Have fields common to both datasets, such as a unique identifier
- Align data elements and values


> Using informatics solutions to merge or join data

- Not all informatics solutions are intended to support merging or joining data sets. Using Excel to merge two data sets is difficult and time consuming. With other tools, like R, Python, you can write instructions to tell the software how to join the datasets you’re working with.

#### 6.2. Types of Merge Operations

#### Joins
Joins are usually used to add information to a data frame: new information is only added to rows in the baseline data frame.


##### Left Join

> The first dataframe is the baseline.

> Left join includes all rows from the left(first) dataset and matched rows from the right dataset. Unmatched rows from the right dataset will have Null values

<img src="https://drive.google.com/uc?export=view&id=1HhW2bMJr8Z_CI2-JcAN4szLxb-jslis2" width=900 height=300/>

```r
# loading Notification data

nd_df <- read_excel('2023 Notification Data.xls', sheet = 1)
head(nd_df)
dim(nd_df)
```

```r
# Loading patient data

yf_df <- read_excel('2023 YF Investigation Data.xls', sheet = 1)
head(yf_df)
dim(yf_df)
```

```r
left_joined_df<- left_join(nd_df, yf_df, by = "NMC Case ID")
dim(left_joined_df)
```

- nd_df: The left DataFrame.
- yf_df: The right DataFrame.
- by='common_column': The column to join on.

If nd_df contains rows not present in yf_df, these rows will be included in left_joined_df, with NaN values for columns from yf_df where no match is found.

```r
head(left_joined_df)
```

##### Right Join

> The second dataframe is the baseline.

> Right join includes all rows from the right dataset and matched rows from the left dataset. Unmatched rows from the left dataset will have Null values

<img src="https://drive.google.com/uc?export=view&id=1MU6jqcvclz-_u9IstE-HbDjXsIWuTj27" width=900 height=300/>

```r
right_joined_df <- right_join(nd_df, yf_df, by = "NMC Case ID")
```

- If yf_df contains rows not present in nd_df, these rows will be included in right_joined_df, with NaN values for columns from nd_df where no match is found.

```r
head(right_joined_df)
```

```r
dim(right_joined_df)
```

##### Inner Join

> Inner join includes only the matching rows from both datasets. Unmatched rows from both datasets are discarded.

```r
inner_merged_df <- merge(nd_df, yf_df, by = "NMC Case ID")
```

```r
inner_joined_df <- inner_join(nd_df, yf_df, by = "NMC Case ID")
```

```r
head(inner_joined_df)
```

```r
dim(inner_joined_df)
```
